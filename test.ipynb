```python
classes = ['greeting', 'goodbye', 'sakit', 'izin', 'general']
pred  = 4
choice = classes[pred]
for i in range(len(data['intents'])):
    if data['intents'][i]['tag'] == choice: 
        index_res = random.randint(0, 2)
        print(data['intents'][i]['responses'][index_res])
```
